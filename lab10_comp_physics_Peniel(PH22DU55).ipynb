## 1. Solve the following boundary values problem using the shooting method:

```python
def Backward_Euler(x1, x2, h, t):
    return np.array([((1-h)*x1+h*x2+h**2*(np.exp(t)-3*np.sin(t)))/(1-h+h**2),(x2+h*np.exp(t)-3*h*np.sin(t)-h*x1)/(1-h+h**2)])


h=0.05
tPlot = np.arange(1, 2+h, h)
x11 = np.zeros(len(tPlot))
x21 = np.zeros(len(tPlot))
x11[0]=1.09737491
x21[0]=0.
for i in range(1, len(tPlot)):
    x11[i] = Backward_Euler(x11[i-1], x21[i-1], h, tPlot[i])[0]
    x21[i] = Backward_Euler(x11[i-1], x21[i-1], h, tPlot[i])[1]


x12 = np.zeros(len(tPlot))
x22 = np.zeros(len(tPlot))
x12[0]=1.09737491
x22[0]=1.
for i in range(1, len(tPlot)):
    x12[i] = Backward_Euler(x12[i-1], x22[i-1], h, tPlot[i])[0]
    x22[i] = Backward_Euler(x12[i-1], x22[i-1], h, tPlot[i])[1]

lam = (8.63749661-x12[len(tPlot)-1])/(x11[len(tPlot)-1]-x12[len(tPlot)-1])
```

```python
plt.plot(tPlot, lam*x11+(1-lam)*x12, '-o', label = 'numerical solution',)
plt.plot(tPlot, np.exp(tPlot)-3*np.cos(tPlot), label='analytical solution')
plt.xlabel('t')
plt.ylabel('x')

#ax[1].set_xlabel('x')
#ax[1].set_ylabel('v')
plt.title('Shooting method using Backward Euler method')
plt.legend()
```

## 2. Write a program that solves the time independent Schrodinger equation: 

```python
from numpy import linalg as LA
h=0.05
xPlot = np.arange(-10, 10+h, h)
```

```python
n= len(xPlot)
A = np.zeros((n, n))
for i, row in enumerate(xPlot):
    if row >= -5 and row<= 5:
        A[i, i] = 1/(h**2) -20
    else:
        A[i, i] = 1/(h**2)
        for j in range(n):
            if i==j-1 or i==j+1:
                A[i, j] = -1/(2*h**2)
A[0, 0]=1
A[0,1]=0
A[n-1, n-1]=1
A[n-1, n-2]=0.
print('The 05 first eigen values are', LA.eig(A)[0][0:5])
fig, ax = plt.subplots(nrows=1, ncols=5, figsize = (30, 4))
for i in range(5):
    ax[i].plot(xPlot, LA.eig(A)[1][:,i])
    ax[i].set_xlabel('x')
    ax[i].set_ylabel('$\psi(x)$')
    ax[i].set_title('eigen state '+str(i))
```
