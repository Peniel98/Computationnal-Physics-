## Finding the root of the function $f(x) = e^{-x} - 5x$

### a) using the bissection method 

```python
import numpy as np 
import matplotlib.pyplot as plt 
def func(x):
        return np.exp(-x) - 5 * x
    
def bissection(a, b):
    i = 0
    list_errors = []
    rel_error = 1.0
    while rel_error > 1e-7:
        c = (a + b)/2
        if (func(a) * func(c) < 0.0):
                rel_error = abs((c-b)/c)
                b = c
                list_errors.append(rel_error)
        elif (func(b) * func(c) < 0.0):
                rel_error = abs((c-a)/c)
                a = c
                list_errors.append(rel_error)
        i+=1
    plt.figure()
    plt.plot(np.arange(0, len(list_errors), 1), list_errors)
    plt.xlabel('nombre of iterations')
    plt.ylabel('relative errors')
    plt.title('relative error vs nombre of iterations')
    return c

bissection(0, 2)
```

### b) Using False position method

```python
def false_position(x1, x):
    i=0
    list_errors = []
    rel_error = 1.0
    while rel_error > 1e-7:
        x0 = x
        x = x1 - (x0 - x1) * func(x1)/(func(x0) - func(x1))
        rel_error = abs((x-x0)/x)
        list_errors.append(rel_error)
        i+=1
    plt.figure()
    plt.plot(np.arange(0, len(list_errors), 1), list_errors)
    plt.xlabel('nombre of iterations')
    plt.ylabel('relative errors')
    plt.title('relative error vs nombre of iterations')
    return x

false_position(0, 2)
```

## c) Newtown-Raphson method 

```python
def func(x):
    return np.exp(-x) - 5 * x
def deriv_func(x):
    return -np.exp(-x) - 5
    
def newton_raphson(x):
    i=0
    x= 2
    list_errors = []
    rel_error = 1.0
    while rel_error > 1e-7:
        x0 = x
        x = x0 - (func(x0)/deriv_func(x0))
        rel_error = abs((x-x0)/x)
        list_errors.append(rel_error)
        i+=1
    plt.figure()
    plt.plot(np.arange(0, len(list_errors), 1), list_errors)
    plt.xlabel('nombre of iterations')
    plt.ylabel('relative errors')
    plt.title('relative error vs nombre of iterations')
    return x

newton_raphson(2)
```
