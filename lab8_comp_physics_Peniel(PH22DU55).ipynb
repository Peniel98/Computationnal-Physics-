```python
import numpy as np 
import matplotlib.pyplot as plt
```

## 1. a) Solve the initial value problems Using Euler, Raltson and Classical fourth order RK method

```python
def func1(x):
    return -2 * x **3 + 12 * x ** 2 - 20 * x + 8.5
def analytic1(x):
    return -x ** 4/2 + 4 * x ** 3 -10 * x ** 2 +8.5 * x + 1
```

```python
def Euler1(h):
    int = np.arange(0., 3.+h, h)
    y = [1.]
    for i in range(len(int)-1):
        y.append(y[i] + func1(int[i]) * h)
    return int, y 

steps = [0.5, 0.25, 0.1, 0.05]
for i in steps:
        plt.plot(Euler1(i)[0], Euler1(i)[1], label = 'h='+str(i))

xPlot = np.linspace(0., 3., 50)
plt.plot(xPlot, analytic1(xPlot), label = 'Analytic')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Euler method')
plt.legend()
```

* Raltson method

```python
def Raltson(h):
    int = np.arange(0., 3.+h, h)
    y = [1.]
    for i in range(len(int)-1):
        y.append(y[i] + (1/3) * h *func1(int[i]) + (2/3) * h * func1(int[i] + (3/4) * h))
    return int, y 

steps = [0.5, 0.25, 0.1, 0.05]
for i in steps:
        plt.plot(Raltson(i)[0], Raltson(i)[1], label = 'h='+str(i))

plt.plot(xPlot, analytic1(xPlot), label = 'Analytic')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Raltson method')
plt.legend()
```

* RK4 method

```python
def RK4(h):
    int = np.arange(0., 3.+h, h)
    y = [1.]
    for i in range(len(int)-1):
        k1 = func1(int[i])
        k2 = func1(int[i]+h/2)
        k3 = func1(int[i]+h/2)
        k4 = func1(int[i]+h/2)
        y.append(y[i] + (h/6) * (k1 + 2 * k2 + 2 * k3 + k4))
    return int, y 

steps = [0.5, 0.25, 0.1, 0.05]
for i in steps:
        plt.plot(RK4(i)[0], RK4(i)[1], label = 'h='+str(i))

plt.plot(xPlot, analytic1(xPlot), label = 'Analytic')
plt.xlabel('x')
plt.ylabel('y')
plt.title('RK4 method')
plt.legend()
```

## 1. b)

```python
def func2(x, y1):
    return -25 * y1 + 25 * np.sin(x) + np.cos(x)
def analytic2(x):
    return np.sin(x) + np.exp(-25 * x)
```

```python
def Euler2(h):
    int = np.arange(0., 6.+h, h)
    y = [1.]
    for i in range(len(int)-1):
        y.append(y[i] + func2(int[i], y[i]) * h)
    return int, y 

steps = [0.5, 0.25, 0.1, 0.05]

fig, ax = plt.subplots(nrows=1, ncols=4, figsize = (15, 2))
xPlot = np.linspace(0, 6, 50)
for i in range(4):
    ax[i].plot(Euler2(steps[i])[0], Euler2(steps[i])[1], label = 'h='+str(steps[i]))
    ax[i].plot(xPlot, analytic2(xPlot), label = 'Analytic')
    ax[i].set_xlabel('x')
    ax[i].set_ylabel('y')
    ax[i].set_title('Euler method')
    ax[i].legend()
```

```python
def Raltson2(h):
    int = np.arange(0., 6.+h, h)
    y = [1.]
    for i in range(len(int)-1):
        y.append(y[i] + h  * ((1/3) * func2(int[i], y[i]) + (2/3) * func2(int[i] + (3/4) * h, y[i] + (3/4) * h * func2(int[i], y[i]))))
    return int, y 

steps = [0.5, 0.25, 0.1, 0.05]

fig, ax = plt.subplots(nrows=1, ncols=4, figsize = (15, 2))

for i in range(4):
    ax[i].plot(Raltson2(steps[i])[0], Raltson2(steps[i])[1], label = 'h='+str(steps[i]))
    ax[i].plot(xPlot, analytic2(xPlot), label = 'Analytic')
    ax[i].set_xlabel('x')
    ax[i].set_ylabel('y')
    ax[i].set_title('Raltson method')
    ax[i].legend()
```

* RK4 Method

```python
def RK42(h):
    int = np.arange(0., 6.+h, h)
    y = [1.]
    for i in range(len(int)-1):
        k1 = func2(int[i], y[i])
        k2 = func2(int[i]+h/2, y[i] + h * k1/2)
        k3 = func2(int[i]+h/2, y[i] + h * k2/2)
        k4 = func2(int[i]+h, y[i] + h * k3)
        y.append(y[i] + (h/6) * (k1 + 2 * k2 + 2 * k3 + k4))
    return int, y 

steps = [0.5, 0.25, 0.1, 0.05]

fig, ax = plt.subplots(nrows=1, ncols=4, figsize = (15, 2))
xPlot = np.linspace(0, 6, 50)
for i in range(4):
    ax[i].plot(RK42(steps[i])[0], RK42(steps[i])[1], label = 'h='+str(steps[i]))
    ax[i].plot(xPlot, analytic2(xPlot), label = 'Analytic')
    ax[i].set_xlabel('x')
    ax[i].set_ylabel('y')
    ax[i].set_title('RK4 method')
    ax[i].legend()
```
