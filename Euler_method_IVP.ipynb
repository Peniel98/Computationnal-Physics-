## Question 2.b 

we first define the useful packages numpy and IVP

```python
import numpy as np 
from IVP import Euler, Heun
```

Next, we define the function of the right hand side of the equation 
$$u'(t)=-3u(t)+2t, u(\pi)=0.$$

```python
def function(u,t):
    return -3*u+2*t
```

Next, we define the function calculatinf the exact solution of the problem 
$$u(t)=\frac23 t -\frac29-\frac23\left(\pi-\frac13\right)e^{(-3(t-\pi))}.$$

```python
def s_exact(t):
        return 2/3*t-2/9-((2/3)*np.pi-2/9)*np.exp(-3*(t-np.pi))
```

We define the initial condition Init=0. The time going from $\pi$ to $0$ and the number of steps Nsteps=15. 

```python
Init = 0
Time = [np.pi, 0]
Nsteps = 15
```

```python
import matplotlib.pyplot as plt
```

```python
for method in [Euler, Heun]:
        u, t = method(function, Time, Init, Nsteps)

        tt = np.linspace(0, np.pi, 101)
        ex = s_exact(tt)
        plt.plot(t, u,'o-',   label='numerical')
        plt.plot(tt, ex,'r-', label='exact')
        plt.title(method.__name__+': number of steps = 15')
        plt.grid(True)
        plt.legend(loc='best')
        plt.show()
```

**Comments**: We clearly see that for a number of steps equal to 15, the Heun's method is much closer to the exact solution than the method of Euler.
